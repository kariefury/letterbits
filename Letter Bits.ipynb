```python
import numpy as np
from scipy.io import wavfile
myFiles = ['a','b','c','d','e','f']
```

```python
i = 0
letterSounds = []
while i < len(myFiles):
    sr, samples = wavfile.read(myFiles[i]+'.wav')
    letterSounds.append(samples)
    i += 1
```

```python
for sound in letterSounds:
    for each in sound:
        print np.binary_repr(each,16)
```
